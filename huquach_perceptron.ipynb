### Problem 1

```python
class RegularPolygon:
    def __init__(self, n, s):
        self.n = n
        self.s = s
    def perimeter(self): # a method .perimeter() which returns the perimeter of the polygon
        return self.n * self.s
    def area(self): # a method .area() which returns the area of the polygon
        radius = self.s/(2*np.tan(np.pi/self.n))
        print("Inradius:", end=" ")
        print(radius)
        print("Area:", end=" ")
        return (self.n/2) * self.s * radius
    def angle(self): # a method .angle() which returns the size (in radians of each angle)
        return (self.n-2)*np.pi/self.n
    def plot(self): # a method .plot() which produces a plot of the polygon
        myangles = np.linspace(0, 2*np.pi, self.n+1)
        radii = self.s/(2*np.sin(np.pi/self.n))
        print("Circumradius:", end=" ")
        print(radii)
        myradii = np.full(self.n+1, radii) 
        ax = plt.subplot(projection='polar')
        ax.plot(myangles, myradii)
        ax.axis('off')
        plt.show()
mypolygon = RegularPolygon(8, 4)
print("Perimeter:", end=" ")
print(mypolygon.perimeter())
print(mypolygon.area())
print("Angle:", end=" ")
print(mypolygon.angle())
mypolygon.plot()
```

### Problem 2

```python
x = np.linspace(0,5,500)
y = np.linspace(0,5,500)

X,Y = np.meshgrid(x,y)
Z = X**2 - 6*X + Y**2 - 4*Y +13
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
ax.view_init(0, -10, 0)
```

The minimum of the function $f(x,y) = x^2 - 6x + y^2 - 4 y + 13$ occurs when $f_{x}=0$ and $f_{y}=0$, and that gives the point $(x,y)=(3,2)$.

```python
def pd(x, y):
    # return partial derivatives
    return np.array([2*x - 6, 2*y - 4])

def graddescent(point, n, eta):
    """Performs gradient descent to find a local minimum of f."""
    for i in range(n):
        gradient = pd(*point)  # Unpack point into x and y
        point = point - eta * gradient
        print(f"Estimated minimum {i+1}: {point}")
    return point
```

```python
start = np.array([2,4])
graddescent(start, 100, 0.1)
```

With $n=100$ and $\eta=0.1$, the graddescent converges to the minimum at $(x,y)=(3,2)$.

```python
graddescent(start, 20, 0.01)
```

With $n=100$ and $\eta=0.1$, the graddescent does not converge to the minimum $(x,y)=(3,2)$.

### Problem 3

```python
data = np.genfromtxt('jan8separate.csv',delimiter=',')
data
```

```python
X=data[:,0:2]
y = data[:,2]
y = np.where(y <= 0, 0, 1)
```

```python
class Perceptron:
    """Perceptron classifier.

    Parameters
    ------------
    eta : float
      Learning rate (between 0.0 and 1.0)
    n_iter : int
      Passes over the training dataset.
    random_state : int
      Random number generator seed for random weight
      initialization.

    Attributes
    -----------
    w_ : 1d-array
      Weights after fitting.
    b_ : Scalar
      Bias unit after fitting.
    errors_ : list
      Number of misclassifications (updates) in each epoch.

    """
    def __init__(self, eta=0.01, n_iter=50, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state

    def fit(self, X, y):
        """Fit training data.

        Parameters
        ----------
        X : {array-like}, shape = [n_examples, n_features]
          Training vectors, where n_examples is the number of examples and
          n_features is the number of features.
        y : array-like, shape = [n_examples]
          Target values.

        Returns
        -------
        self : object

        """
        rgen = np.random.RandomState(self.random_state)
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=X.shape[1])
        self.b_ = np.float_(0.)
        
        self.errors_ = []

        for _ in range(self.n_iter):
            errors = 0
            for xi, target in zip(X, y):
                update = self.eta * (target - self.predict(xi))
                self.w_ += update * xi
                self.b_ += update
                errors += int(update != 0.0)
            self.errors_.append(errors)
        return self

    def net_input(self, X):
        """Calculate net input"""
        return np.dot(X, self.w_) + self.b_

    def predict(self, X):
        """Return class label after unit step"""
        return np.where(self.net_input(X) >= 0.0, 1, 0)
```

```python
model = Perceptron(0.1, 100, 0)
model.fit(X,y)
print(model.predict(X) == y)
print(np.count_nonzero(model.predict(X)-y))
```
